## 동전 분배하기

n개의 동전을 a, b, c 세 명에게 나누어 주려고 하는데, 세 명이 받은 각각의 총액을 계산해 총액이 가장 큰 사람과 가장 작은 사람의 차가 최소가 되도록 최소차를 출력하시오.

```python
n = int(input())
lst = [int(input()) for _ in range(n)]
```

```python
abc = [0]*3
min_dif = 200000000
def DFS(x):
    global min_dif
    if x == n:
        print(abc)
        max_num = max(abc)
        min_num = min(abc)
        dif = max_num - min_num
        if dif < min_dif and len(set(abc)) == 3:
            min_dif = dif
            print("min_dif는 : ", min_dif)
        
    else:
        for i in range(3):
            abc[i] += lst[x]
            DFS(x+1)
            abc[i] -= lst[x]
        
```

```python
DFS(0)
print(min_dif)
```
