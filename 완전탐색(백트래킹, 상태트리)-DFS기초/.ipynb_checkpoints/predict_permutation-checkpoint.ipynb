## 수열 추측하기

가장 윗줄에 1부터 n까지의 숫자가 한 개 씩 적혀져 있고, 파스칼의 삼각형처럼 위의 두 개를 더한 값이 저장되어 최종적으로는 마지막줄에 숫자 하나가 들어간다.(역삼각형)

n과 맨 마지막 줄의 숫자가 주어지면, 맨 윗줄의 숫자들을 구하는 프로그램을 작성하시오.

```python
n, f = map(int, input().split())
```

```python
res = [0]*n
def DFS(x):
    if x == n: #4개 뽑으면 그 숫자들로 더해내려갔을 때 16이 나오면 res 출력
        for i in res:
             print(i, end = ' ')
        print()
        
#         if pas(n) == f:
#             print()
#             print(' '.join(res))
#         else:
#             print('NO')
    else:
        for i in range(1, n+1):
            res[x] = i
            print(res)
            DFS(x+1)
```

```python
sums = res
def pas(l):
    print(sums)
    print(l)
    if l == 1:
        return sums[0]
    else:
        lenn = len(sums) #4
        for i in range(1, lenn): #1, 2, 3
            tmp = sums[0] + sums[1]
            sums.append(tmp)
            sums.pop(0)
            
        else:
            sums.pop(0)
            pas(l-1)
```
