## 창고정리

상자가 가로 방향으로 일렬로 쌓여 있다.
창고 높이 조정은 가장 높은 곳에서 상자를 가장 낮은 곳으로 이동하는 것을 말한다.

가장 높은 곳이나 가장 낮은 곳이 여러 곳이면 아무거나 선택

m회의 높이 조정 후 가장 높은 곳과 가장 낮은 곳의 차이를 출력하는 프로그램 작성

```python
10
69 42 68 76 40 87 14 65 76 81
50
```

```python
l = int(input())
hei_lst = list(map(int,input().split()))
m = int(input())
```

```python
def check(hei_lst):
    maxh = max(hei_lst)
    minh = min(hei_lst)
    go1 = True
    go2 = True
    for i in range(len(hei_lst)):
        if hei_lst[i] == maxh and go1 == True:
            hei_lst[i] -= 1
            go1 = False
            
        elif hei_lst[i] == minh and go2 == True:
            hei_lst[i] += 1
            go2 = False
    return hei_lst

for i in range(m):
    hei_lst = check(hei_lst)
    #print(hei_lst)

print(max(hei_lst)-min(hei_lst))
```
